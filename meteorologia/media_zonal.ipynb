```python
import xarray as xr
import matplotlib.pyplot as plt
```

```python
ds = xr.open_dataset('../dados/NetCDF/precip.mon.mean.nc')
```

```python
ds
```

```python
ds.precip.mean(dim=('lon', 'time')).plot(label='Clima', color='black')
ds.precip.sel(time=slice('2019-12-01', '2020-02-01')).mean(dim=('lon', 'time')).plot(label='DJF', color='blue')
ds.precip.sel(time=slice('2020-06-01', '2020-08-01')).mean(dim=('lon', 'time')).plot(label='JJA', color='green')

plt.legend()
```
