```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../dados/texto/numero_dias_sem_chuva.txt', sep=' ')
```

```python
df
```

```python
df = df.set_index('Ano')
```

```python
acumulado = df.cumsum(axis=1)
```

```python
acumulado
```

```python
acumulado.columns
```

```python
acumulado.index
```

```python
clima = acumulado.mean().round(0)
```

```python
clima
```

```python
plt.plot(acumulado.columns, acumulado.loc[2019], label='2019', color='r')
plt.plot(acumulado.columns, acumulado.loc[2020], label='2020', color='b')
plt.plot(acumulado.columns, clima, label='Média', color='y')
plt.legend(frameon =False, loc='lower right', fontsize=8)
```
